# Load Power poll

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_excel('Power2.xlsx')
```

# Sort by Appeal

```python
df.sort_values(['Your Appeal'], ascending=False)
```

### Where am I strong?

```python
df3=df.loc[(df['Your Appeal']>0.50)]
outcome1=df3.sort_values(['Your Appeal'], ascending=False)
outcome1
```

### Where am I weak?

```python
df4=df.loc[(df['Your Appeal']<0.20)]
outcome2=df4.sort_values(['Your Appeal'], ascending=False)
outcome2
```

# Population shares sorting

```python
df2 = df.drop(columns=['Estimated Turnout', 'Special Effects'])
df2 = df [['Population Share', 'Demographic','Your Appeal','Social Position','Economic Position']]
df2 = df2.reset_index(drop=True)
result = df2.sort_values(['Population Share'], ascending=False)
result
```

# Social Position breakdown

```python
df['count'] = 1
c=df.groupby(['Social Position']).count()['count']
print(c)
plt.figure(figsize=(6,4), dpi=120)
plt.title('Social Position distribution')
plt.xlabel('Social Position')
plt.ylabel('Occurrences')
c.plot.bar()
plt.show()


#plt.bar(range(len(c)), c)
```

```python
c.plot.pie(autopct='%.2f')
plt.title("Pie Chart of Social Position distribution")
plt.show()
```
